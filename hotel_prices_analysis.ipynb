```python
import pandas as pd
import matplotlib
import numpy as np

# Import dataset.
df = pd.read_csv(filepath_or_buffer='sample_data.csv', sep=';', decimal=',')

# Convert to proper dateformat.
df['booking date'] = pd.to_datetime(df['booking date'], format = '%d.%m.%Y')
df['date'] = pd.to_datetime(df['date'], format = '%d.%m.%Y')
```

```python
# Draw a plot that shows the development of the average hotel price.
ax = df.plot(kind='line', x='date', y='mean')

# Consider adding some dates with empty values afterwards here for display.
```

```python
# Find the lowest avg. price and its date.
lowest_price = df['mean'].min()
lowprice_date = df.loc[df['mean'] == lowest_price, 'date']

# Calculate the days left until the booking date.
booking_date = df.loc[df['mean'] == lowest_price, 'booking date']
# Note: I'm not sure the slicing here is consistent.
# See what happens when you have more data.
days_left = int(str(booking_date - lowprice_date)[5:-28])

# Output the result.
# Add logic to convert to weeks, months, if you have more than 31 days.
print(f"The cheapest average hotel price was {lowest_price} €, recorded " +
      f"{days_left} days before the booking date.")
```
